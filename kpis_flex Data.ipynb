# Data Preprocessing

```python
# Loading functions
from pandas import read_csv
from pandas import to_datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the csv file, changing the decimal character and recognizing
# missing values
kpis1 = read_csv('kpis_flex1.csv', decimal=',', na_values = '!NULL!')
```

```python
# Convert column time to datetime 
kpis1['time'] = to_datetime(kpis1['time'], infer_datetime_format=True)

# Set time as index (good practice for time series data)
kpis1 = kpis1.set_index(kpis1.time)

# Drop (remove) the old column time
kpis1 = kpis1.drop(['time'], axis=1)
```

```python
# Check the dataframe
kpis1.head()
```

```python
# check data types
kpis1.dtypes
```

```python
# Reading the csv file and changing the character for decimals.
kpis2 = read_csv('kpis_flex2.csv', decimal=",", na_values = '!NULL!')
```

```python
# Convert column time to datetime 
kpis2['time'] = to_datetime(kpis2['time'], infer_datetime_format=True)

# Set time as index (good practice for time series data)
kpis2 = kpis2.set_index(kpis2.time)

# Drop (remove) the old column time
kpis2 = kpis2.drop(['time'], axis=1)
```

```python
# Check the dataframe
kpis2.head()
```

```python
# check data types
kpis2.dtypes
```
